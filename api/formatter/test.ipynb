```python
import math
import json
import numpy as np
import pandas as pd
```

```python
class JsonFormatter:
    def __init__(self, file_path):
        self.file_path = file_path

    def format(self):
        df = pd.read_csv(self.file_path)
        df = df[["id", "name", "type1", "type2", "height", "weight", "status", "hp", "attack", "defence", "sp_atk", "sp_def", "speed"]]
        df.loc[df["type2"].isnull() , "types"] = df["type1"]
        df.loc[~df["type2"].isnull() , "types"] = df["type1"] + " " + df["type2"]

        return df
```

```python
formatter = JsonFormatter("../data/pokemon.csv")
df = formatter.format()
df
```
